## Introduction to Audio Tokenization

When we talk about tokenizing audio for machine learning, we’re generally looking for a way to transform a continuous waveform into a sequence of discrete symbols. This is analogous to how words or subwords are used in text. The benefits include:

- **Reduced data size:** Instead of dealing with raw waveforms or high-dimensional embeddings, you have discrete tokens that can be stored and processed more easily.
- **Compatibility with text-based models:** Many neural architectures for NLP expect sequences of discrete tokens. Mapping audio to a token sequence can let you reuse these architectures.
- **Abstraction of acoustic features:** Tokens can capture phonetic or acoustic characteristics in a way that generalizes across different speakers or recording conditions.

[HuBERT](https://arxiv.org/abs/2106.07447) is a self-supervised model that learns robust speech representations. After the embeddings are extracted, we apply a clustering algorithm such as K-Means to quantize the embeddings into discrete tokens.

## Loading and Preprocessing the Audio

Below is a snippet of Python code that shows how we load a local MP3 file into a [Hugging Face `Dataset`](https://huggingface.co/docs/datasets). We then cast the dataset’s `audio` column to the `Audio` feature type, which automatically decodes MP3 into raw waveforms. After that, we make sure the waveform is a PyTorch tensor and resample it to 16 kHz if needed.

````python
import time
start_time = time.time()

import torch
import torchaudio
import matplotlib.pyplot as plt
import torchaudio.transforms as T
from transformers import Wav2Vec2Processor, HubertModel
from datasets import Dataset, Audio
import numpy as np
from sklearn.cluster import KMeans

# Auto-detect device
device = "cuda" if torch.cuda.is_available() else ("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

# Example: single MP3 file path
data = {"audio": ["/Users/badeepak/Downloads/Arthur.mp3"]}

# Build dataset from a Python dictionary
dataset = Dataset.from_dict(data)

# Cast 'audio' column to automatically decode MP3
dataset = dataset.cast_column("audio", Audio(decode=True))

# Access the decoded audio data
audio_info = dataset[0]["audio"]
waveform = audio_info["array"]  # A NumPy float32 array
sample_rate = audio_info["sampling_rate"]

# Convert waveform to a PyTorch tensor
waveform = torch.from_numpy(waveform).float()

# Ensure shape is (channels, time)
if waveform.ndim == 1:
    waveform = waveform.unsqueeze(0)

# Resample if the sample rate isn't 16 kHz
if sample_rate != 16000:
    resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
    waveform = resampler(waveform)
    sample_rate = 16000

print(f"Waveform shape: {waveform.shape}")
print(f"Sample rate: {sample_rate}")
````

## Visualizing the Audio as a Spectrogram

To gain insight into our audio data, we can visualize a Mel spectrogram using `torchaudio.transforms.MelSpectrogram`. The spectrogram shows how energy is distributed across different frequency bins over time.

````python
def visualize_spectrogram(waveform, sample_rate, title="Spectrogram", duration=None):
    """
    Plots the mel spectrogram of a waveform.

    Parameters:
      waveform: 2D tensor (channels x time)
      sample_rate: Audio sample rate
      title: Plot title
      duration: if provided, only the first `duration` seconds of the waveform are visualized.
    """
    # If duration is specified, slice the waveform to the first `duration` seconds
    if duration is not None:
        num_samples = int(duration * sample_rate)
        waveform = waveform[:, :num_samples]

    # If needed, convert to single-channel for plotting
    if waveform.ndim > 2:
        waveform = waveform[0].unsqueeze(0)

    # Create the mel spectrogram
    mel_spectrogram = T.MelSpectrogram(sample_rate=sample_rate, n_mels=128)
    mel = mel_spectrogram(waveform)  # shape: (1, 128, time_frames)

    # Convert to decibels
    mel_db = mel.log2()[0].numpy()  # shape: (128, time_frames)

    # Plot
    plt.figure(figsize=(10, 4))
    plt.imshow(mel_db, aspect='auto', origin='lower')
    plt.title(title)
    plt.colorbar(format="%+2.0f dB")
    plt.xlabel("Time")
    plt.ylabel("Mel Frequency")
    plt.show()

# Visualize the waveform as a spectrogram
visualize_spectrogram(waveform, sample_rate, title="Spectrogram (First 5 Seconds)", duration=5)
````

## Visualizing the Waveform (First 5 Seconds)

Below, we visualize the raw audio waveform for the first 5 seconds. This plot shows how the amplitude of the signal changes over time.

````python
# Visualize waveform for the first 5 seconds
def plot_waveform_segment(waveform, sample_rate, duration=5, title="Waveform (First 5 Seconds)"):
    """
    Plots the waveform for the first `duration` seconds.
    """
    import matplotlib.pyplot as plt
    # Calculate the number of samples for the desired duration
    num_samples = int(duration * sample_rate)
    waveform_segment = waveform[:, :num_samples]
    # Create a time axis from 0 to duration seconds
    time_axis = torch.linspace(0, duration, waveform_segment.shape[1])
    plt.figure(figsize=(10, 4))
    plt.plot(time_axis.cpu().numpy(), waveform_segment.squeeze().cpu().numpy())
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title(title)
    plt.show()

plot_waveform_segment(waveform, sample_rate)
````

## Extracting Embeddings and Transcribing Audio with HuBERT

[HuBERT](https://huggingface.co/facebook/hubert-large-ls960-ft) is a self-supervised model that learns robust speech representations. We’ll show two uses:

1. **Extracting embeddings** (a continuous representation of the audio).
2. **Transcribing** the audio to text by using a model with a CTC head.

````python
# 1) Generate embeddings with the base HuBERT model
processor_base = Wav2Vec2Processor.from_pretrained("facebook/hubert-large-ls960-ft")
model_base = HubertModel.from_pretrained("facebook/hubert-large-ls960-ft")

def tokenize_audio(waveform, sample_rate):
    """
    Runs HuBERT model on audio and returns the last hidden state (token embeddings).
    """
    # Move waveform to device
    waveform = waveform.to(device)

    # HuBERT's processor needs NumPy input
    inputs = processor_base(
        waveform.squeeze().cpu().numpy(),
        sampling_rate=sample_rate,
        return_tensors="pt"
    )

    with torch.no_grad():
        outputs = model_base(**inputs)
    return outputs.last_hidden_state  # shape: (batch_size=1, time, 1024)

tokens = tokenize_audio(waveform, sample_rate)
print("Tokenized Audio Shape:", tokens.shape)
````

````python
# 2) Transcribe audio with a HuBERT model that has a CTC head
from transformers import HubertForCTC

processor_ctc = Wav2Vec2Processor.from_pretrained("facebook/hubert-large-ls960-ft")
model_ctc = HubertForCTC.from_pretrained("facebook/hubert-large-ls960-ft").to(device)

def transcribe_audio(waveform, sample_rate):
    """
    Runs HuBERT model on audio and returns transcribed text.
    """
    # Move waveform to device
    waveform = waveform.to(device)

    # Convert to CPU-based NumPy for the processor
    inputs = processor_ctc(
        waveform.squeeze().cpu().numpy(),
        sampling_rate=sample_rate,
        return_tensors="pt"
    )
    # Move inputs to device
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        logits = model_ctc(**inputs).logits

    predicted_ids = torch.argmax(logits, dim=-1)
    transcription = processor_ctc.batch_decode(predicted_ids)[0]
    return transcription

transcription = transcribe_audio(waveform, sample_rate)
print("Transcription:", transcription)
````

## Discretizing Embeddings with K-Means

The extracted HuBERT embeddings are still continuous vectors of size 1024 at each time step. To turn these into discrete tokens, we cluster them with K-Means. Below, we create 100 clusters, meaning each frame of audio will be assigned to one of 100 discrete tokens.

````python
# Extract embeddings (if not already)
# 'tokens' is shape (1, T, 1024)
embeddings_np = tokens.squeeze(0).cpu().numpy()  # shape: (T, 1024)

# Perform K-Means clustering on the embeddings with 100 clusters
n_clusters = 100
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(embeddings_np)
hubert_tokens = kmeans.predict(embeddings_np)

print("HuBERT Token IDs:", hubert_tokens[:10])  # print first 10 tokenized values
print("Tokenized Shape:", hubert_tokens.shape)  # e.g. (T,)
````
